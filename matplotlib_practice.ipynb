```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add the label for the legend
plt.plot(x_axis, y_axis, marker = "*", color="blue", linewidth=2, label= 'Boston')
# Create labels for x and y axis
plt.xlabel('Date')
plt.ylabel('Fare($)')
# Set the y limit between 0 and 45
plt.ylim(0,45)
# Create a title
plt.title('Pyber Fare by Month')
# Add a grid
plt.grid()
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker = "D", color = "green", linewidth = 2, label = "Boston")
# Set the labels
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
# Set Y limit between 0 and 45
ax.set_ylim(0,45)
# Create a title
ax.set_title("Object Oriented Pyber by Month")
# Add a gird
ax.grid()
# Add a legned
ax.legend()
```
